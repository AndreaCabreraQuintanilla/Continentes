```python
# Tratamiento de datos
# -----------------------------------------------------------------------
import numpy as np
import pandas as pd

# Estadísticos
# ------------------------------------------------------------------------------
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.multivariate.manova import MANOVA
from sklearn.preprocessing import StandardScaler
```

Hasta ahora habéis estado evaluando las características de vuestro set de datos y habéis hecho una gran exploración, es el momento de hacer vuestro primer ANOVA! En el ejercicio de hoy tendréis que hacer un ANOVA con vuestro datos y hacer una interpretación de los resultados.
📌 NOTA Puede que vuestros datos no se ajusten o no cumplan todas las asunciones, no pasa nada, haced el ANOVA e interpretad los resultados. En próximas lecciones aprenderemos que podemos hacer cuando nos encontramos en esta situación.

```python
df = pd.read_pickle("../data/6_estandarizacion.pkl")
df.sample(5)
```

```python
lm = ols('salario_box ~ cesta_compra + bebidas_super + bebidas_rest + comida_rest  + ropa + deportes + cine + marlboro + educacion + transporte + gasolina + vehiculos + internet + movil + gastos_fijos + alquiler + m2 + inte_fij_hipo', data=df).fit()
sm.stats.anova_lm(lm)
```

```python
lm.summary()
```

En el caso de que pudiésemos utilizar ANOVA, obtendríamos que nuestra variable respuesta "salario" estaría explicada en un 63% por las variables predictoras. Por otro lado, las variables que mejor predicen nuestra variable respuesta son "cesta_compra" con 
4889.633762 en F y "bebidas_rest" con 1603.893688 en F.
